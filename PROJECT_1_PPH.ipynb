## Memuat data

### Subtask:
Memuat data dari file CSV yang disediakan ke dalam DataFrame pandas.


**Reasoning**:
The first step is to import the pandas library and load the data from the specified CSV file into a pandas DataFrame. Then, display the first five rows to verify the data loading.



```python
import pandas as pd

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('/content/laporan_akuntansi_pajak_integrasi_2023_2027 (1).csv')

# Display the first five rows of the DataFrame
display(df.head())
```

The data is loaded, but the columns are not parsed correctly, it seems the delimiter is semicolon instead of comma. I need to load the data again using the correct delimiter.



```python
# Load the CSV file into a pandas DataFrame with semicolon delimiter
df = pd.read_csv('/content/laporan_akuntansi_pajak_integrasi_2023_2027 (1).csv', delimiter=';')

# Display the first five rows of the DataFrame
display(df.head())
```

## Pembersihan dan persiapan data

### Subtask:
Memeriksa dan membersihkan data jika ada nilai yang hilang, format yang tidak sesuai, atau anomali lainnya. Mempersiapkan data untuk analisis dan visualisasi.


**Reasoning**:
Check the general information about the DataFrame, including data types and missing values, to understand its structure and identify initial cleaning needs.



```python
df.info()
```

**Reasoning**:
Examine the last few rows of the DataFrame to identify any irrelevant rows at the end, as suggested by the `df.info()` output showing fewer non-null values towards the end of the DataFrame.



```python
display(df.tail())
```

**Reasoning**:
Based on the `df.tail()` output, the last 12 rows (from index 25 to 36) contain metadata and not the actual financial data. These rows need to be removed before proceeding with the analysis.



```python
df = df.iloc[:25].copy()
display(df.tail())
```

**Reasoning**:
The `df.tail()` output shows that rows 20, 21, 22, 23, and 24 also contain metadata and not the financial data. I need to further remove these rows to keep only the relevant data.



```python
df = df.iloc[:20].copy()
display(df.tail())
```

**Reasoning**:
The tail still shows rows with all NaN values. I need to remove these rows as well to have a clean dataset.



```python
df.dropna(how='all', inplace=True)
display(df.tail())
```

**Reasoning**:
Based on the `df.info()` output, many columns are of type object and contain non-numeric characters like '.' and ','. I need to clean these columns by removing these characters and then convert them to a numeric data type to prepare for analysis. I will apply this cleaning to all relevant columns.



```python
cols_to_clean = [
    'persediaan_awal', 'persediaan_akhir', 'hpp_rawat_inap', 'hpp_rawat_jalan',
    'total_hpp', 'pendapatan_operasional', 'pendapatan_penjualan', 'ppn_keluaran',
    'ppn_masukan', 'ppn_netto', 'laba_kotor_penjualan', 'beban_operasional',
    'laba_kotor_operasional', 'laba_sebelum_pajak', 'laba_kena_pajak',
    'laba_bersih_normal', 'laba_bersih_tax_holiday'
]

for col in cols_to_clean:
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
        df[col] = pd.to_numeric(df[col], errors='coerce')

display(df.info())
display(df.head())
```

## Analisis data awal

### Subtask:
Melakukan analisis deskriptif awal untuk memahami struktur data, ringkasan statistik, dan distribusi variabel kunci.


**Reasoning**:
Display the descriptive statistics for numeric columns, the number of unique values for 'tahun' and 'tax_holiday_status', and the value counts for 'tax_holiday_status' as requested by the instructions to perform initial descriptive analysis.



```python
# Display descriptive statistics for numeric columns
display(df.describe())

# Display the number of unique values for 'tahun' and 'tax_holiday_status'
display(df['tahun'].nunique())
display(df['tax_holiday_status'].nunique())

# Display the value counts for 'tax_holiday_status'
display(df['tax_holiday_status'].value_counts())
```

## Analisis mendalam dan identifikasi tren

### Subtask:
Melakukan analisis yang lebih mendalam untuk mengidentifikasi tren, pola, atau hubungan antar variabel yang relevan dengan tujuan analisis.


**Reasoning**:
Calculate the total net income by summing 'laba_bersih_normal' and 'laba_bersih_tax_holiday', group the data by year, calculate the mean of numeric columns, and then analyze the trends and relationships as requested.



```python
# 1. Hitung laba bersih total (laba_bersih_normal + laba_bersih_tax_holiday)
df['total_laba_bersih'] = df['laba_bersih_normal'] + df['laba_bersih_tax_holiday']

# 2. Kelompokkan data berdasarkan tahun dan hitung rata-rata dari kolom numerik.
df_grouped_by_year = df.groupby('tahun').mean(numeric_only=True)
display(df_grouped_by_year)

# 3. Analisis tren laba bersih total dari tahun ke tahun.
print("\nAnalisis Tren Laba Bersih Total dari Tahun ke Tahun:")
display(df_grouped_by_year['total_laba_bersih'])

# 4. Analisis hubungan antara pendapatan operasional dan laba bersih total.
print("\nAnalisis Hubungan antara Pendapatan Operasional dan Laba Bersih Total:")
display(df_grouped_by_year[['pendapatan_operasional', 'total_laba_bersih']])

# 5. Analisis hubungan antara total HPP dan laba bersih total.
print("\nAnalisis Hubungan antara Total HPP dan Laba Bersih Total:")
display(df_grouped_by_year[['total_hpp', 'total_laba_bersih']])
```

## Visualisasi data

### Subtask:
Membuat grafik berwarna yang relevan berdasarkan analisis data untuk memvisualisasikan temuan kunci. Memilih jenis grafik yang paling sesuai untuk data dan analisis yang dilakukan (misalnya, diagram batang, diagram garis, diagram lingkaran, dll.).


**Reasoning**:
Import the necessary libraries for plotting and create the requested visualizations based on the grouped data, ensuring different colors are used and displaying each plot.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set_style('whitegrid')

# 1. Line plot for total_laba_bersih trend over the years
plt.figure(figsize=(10, 6))
sns.lineplot(x=df_grouped_by_year.index, y='total_laba_bersih', data=df_grouped_by_year, marker='o', color='skyblue')
plt.title('Tren Total Laba Bersih dari Tahun ke Tahun')
plt.xlabel('Tahun')
plt.ylabel('Total Laba Bersih')
plt.xticks(df_grouped_by_year.index)
plt.grid(True)
plt.show()

# 2. Bar plot for the relationship between pendapatan_operasional and total_laba_bersih per year
df_grouped_by_year[['pendapatan_operasional', 'total_laba_bersih']].plot(kind='bar', figsize=(12, 7), color=['lightcoral', 'mediumseagreen'])
plt.title('Pendapatan Operasional vs Total Laba Bersih per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.xticks(rotation=0)
plt.legend(title='Metrik')
plt.grid(axis='y')
plt.show()

# 3. Line plot for the relationship between total_hpp and total_laba_bersih per year
plt.figure(figsize=(10, 6))
sns.lineplot(x=df_grouped_by_year.index, y='total_hpp', data=df_grouped_by_year, marker='o', color='orange', label='Total HPP')
sns.lineplot(x=df_grouped_by_year.index, y='total_laba_bersih', data=df_grouped_by_year, marker='o', color='purple', label='Total Laba Bersih')
plt.title('Hubungan antara Total HPP dan Total Laba Bersih per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Jumlah')
plt.xticks(df_grouped_by_year.index)
plt.legend()
plt.grid(True)
plt.show()
```
